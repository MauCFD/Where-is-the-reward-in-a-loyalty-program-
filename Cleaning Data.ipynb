```python
import os
import pandas as pd
import numpy as np
```

```python
#Read CSV's
location = "data/CT_DB_Location.csv"
location_df = pd.read_csv(location)
points = "data/CT_DB_Points.csv"
points_df = pd.read_csv(points)
sales = "data/CT_DB_Sales.csv"
sales_df = pd.read_csv(sales)


#Combine Data
combined_data = pd.merge(sales_df, points_df, on=["Client ID", "Quarter", "Month", "Year"], how="outer")
combined_data = pd.merge(combined_data, location_df, on= "Client ID", how="outer")
#Read Combine Data
combined_data
```

```python
#Replacing the original values to a new format which we can convert from string to float and get the Totals

#Format to column "Min Monthly Sales"
combined_data['Min Monthly Sales'] = combined_data['Min Monthly Sales'].replace(',','', regex = True)
combined_data['Min Monthly Sales'] = combined_data['Min Monthly Sales'].str.replace('\$', '', regex=True).astype('float')

#Format to Column "A"
combined_data['A'] = combined_data['A'].replace(',','', regex = True)
combined_data['A'] = combined_data['A'].str.replace('\$', '', regex=True).astype('float')

#Format to Column "B"
combined_data['B'] = combined_data['B'].replace(',','', regex = True)
combined_data['B'] = combined_data['B'].str.replace('\$', '', regex=True).astype('float')

#Format to Column "C"
combined_data['C'] = combined_data['C'].replace(',','', regex = True)
combined_data['C'] = combined_data['C'].str.replace('\$', '', regex=True).astype('float')


combined_data
```

```python
# Adding the Total Sales and Points, and Location columns
combined_data["Total Sales"] = combined_data["A"] + combined_data["B"] + combined_data["C"]
combined_data["Total Points"] = combined_data["Monthly Points"] + combined_data["Quarter Bonus 1"] + combined_data["Quarter Bonus 2"] + combined_data["Yearly Bonus 1"] + combined_data["Yearly Bonus 1"] 

# We know that the program estimates each point is worth $0.86 MXN, so we will create a column to know the Points worth in $
combined_data["Points Worth ($)"] = combined_data["Total Points"] * 0.86

#We would like to know the compliance of the client in their monthly sales, so we will add a column that calculates it.
combined_data["Compliance"] = round(combined_data["Total Sales"] / combined_data["Min Monthly Sales"], 2)
combined_data
```

```python
# Reorganizing the data frame
Project_df = combined_data[["Client ID", "Town", "State", "Zip Code", "Month", "Quarter", "Year", "Category", 
                               "Distributor", "Min Monthly Sales", "Total Sales", "Compliance", "Total Points", "Points Worth ($)"]]
Project_df['Total Sales'] = Project_df['Total Sales'].astype('int')
Project_df['Total Points'] = Project_df['Total Points'].astype('int')

# Organized and clean data frame we will be working with
Project_df
```

```python
# Export the Project_df into a csv
Project_df.to_csv("Project_df.csv",encoding="utf-8",index=False)
```
